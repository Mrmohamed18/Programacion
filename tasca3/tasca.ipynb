#### *La función enumerate() se utiliza para convertir una entrada en una enumeración.*

```python
frutas = ["manzana", "banana", "cereza"]
for fruita, fruta in enumerate(frutas):
  print(f"Fruita: {fruita}, Fruta: {fruta}")
```

#### *La funció map() a Python aplica una funció determinada a cada element d'un iterable (llista, tupla, conjunt, etc.) i retorna un objecte mapa (iterador). És una funció d'ordre superior que s'utilitza per a transformacions uniformes element a element, permetent un codi concís i eficient.*

```python
numero = ['1', '2', '3', '4']
num = map(int, numero)
print(list(num))
```

#### *filter() te permite extraer elementos de una lista, tupla o cualquier iterable, pero sólo si superan tu prueba*

```python
def filterr(num):
    return num % 2 == 0

numeros = [1, 2, 3, 4, 5, 6]
resultat = filter(filterr, numeros)
print(list(resultat))
```

#### *reduce() es una funcion que toma como argumento un conjunto de valores (una lista, una tupla, o cualquier objeto iterable) y lo "reduce" a un único valor*

```python
from functools import reduce

numeros = [10, 42, 23, 14, 57]
resultado = reduce(lambda a, b: a + b, numeros)
print(resultado)
```

#### Crea un programa que aleatòriament generi una llista amb nombres desordenats i algun de repetit. Que el programa doni resposta a les següents preguntes:

```python
import random

llista = []
for i in range(20):
    llista.append(random.randint(1, 10))

print("Llista original : ", llista)

print("Quants números hi ha ?: ", len(llista))
print("Quantes vegades apareix el número 3 ? : ", llista.count(3))
print("Quantes vegades apareixen el 3 i el 4 ? : ", llista.count(3) + llista.count(4))
print("Número més gran : ", max(llista))
print("Els 3 números més petits : ", sorted(llista)[:3])
print("Rang de la llista : ", max(llista) - min(llista))

llista_final = sorted(set(llista))
print("Llista final sense repeticions : ", llista_final)
```

## Exercici 5

```python
compra = {
    "Pomes": {"Qty": 5, "€": 0.42},
    "Peres": {"Qty": 3, "€": 0.66},
    "Plàtans": {"Qty": 6, "€": 0.35}
}

cost_peres = compra["Peres"]["Qty"] * compra["Peres"]["€"]
print("Cost total de les Peres:", round(cost_peres, 2), "€")

total_fruites = 0
for dades in compra.values():
    total_fruites += dades["Qty"]
print("Total de fruites comprades:", total_fruites)

fruita_mes_cara = None
preu_max = 0
for fruita, dades in compra.items():
    if dades["€"] > preu_max:
        preu_max = dades["€"]
        fruita_mes_cara = fruita
print("Fruita més cara:", fruita_mes_cara, "amb un preu de", preu_max, "€")

def resum_compra(compra):
    cost_total = 0
    total_unitats = 0
    for dades in compra.values():
        cost_total += dades["Qty"] * dades["€"]
        total_unitats += dades["Qty"]
    preu_mig = cost_total / total_unitats
    return cost_total, preu_mig

cost_total, preu_mig = resum_compra(compra)
print("Cost total de la compra:", round(cost_total, 2), "€")
print("Preu mitjà per fruita:", round(preu_mig, 2), "€")
```

## Exercici 7

#### Escriu el mateix programa, peò amb les variants a, b i c que hi ha a continuació, on s’introdueixin paraules i es generi una llista amb les longituds de cada paraula que s’introdueix. 
#### És a dir, per exemple, el resultat per a les paraules ["hola", "món", "python"]ha de ser [4, 3, 6].

### A-Variant bucle for
#### Construeix amb un bucle for tradicional una llista amb les longituds de cada paraula que s’introdueix (sense utilitzar filter() ni map()).

#### És a dir, el resultat per a les paraules ["hola", "món", "python"]ha de ser [4, 3, 6].

```python
paraules = ["hola", "món", "python"]
longitud = []

for i in paraules:
    longitud.append(len(i))
    
print(longitud)
```

### B-Variant map()
#### Construeix amb map una llista amb les longituds de cada paraula que s’introdueix.

#### És a dir, el resultat per a les paraules ["hola", "món", "python"]ha de ser [4, 3, 6].

```python
paraules = ["hola", "món", "python"]

n = list(map(len, paraules))
print(n)
```

### C-Variant list comprehension
#### Construeix amb list comprehension (sense map() ni list()) una llista amb les longituds de cada paraula que s’introdueix.

#### És a dir, el resultat per a les paraules ["hola", "món", "python"]ha de ser [4, 3, 6].

```python
paraules = ["hola", "món", "python"]

n = [len(paraula) for paraula in paraules ]
print(n)
```

## Exercici 8

#### A partir de l’exercici 7b, utilitzant filter i lambda, queda’t només amb les paraules que comencin per la lletra que s’indica.

#### Per exemple, agafant l’exemple de paraules de l’exercici 7, si s’indica la lletra “p”, el resultat ha de ser ["python"]

```python
paraules = ["hola", "món", "python"]

n = list(filter(lambda palabra: palabra.startswith('p'), paraules))

print(n)
```

#### • Què passaria en l’exercici 8 si hi hagués paraules buides en la llista de paraules (“”) ?
#### Si en el ejercicio 8 se esta utilitzando palabra[0] y la cadena esta vacia falla porque no tiene ninguno caracter y nos da el error(indexError)
#### • Perquè?
####  Porque palabra[0] intenta leer el primer caracter pero si la cadena esta vacia no existe ninguno y el python nos va dar error.
